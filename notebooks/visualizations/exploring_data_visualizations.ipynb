```python
# Check for missing values in the 'primary_focus_subject' column
print(new_df['primary_focus_subject'].isnull().sum())

# Fill missing values with a placeholder (or remove rows with missing values)
new_df['primary_focus_subject'].fillna('Missing', inplace=True)

# Display the distribution
print(new_df['primary_focus_subject'].value_counts())

# Visualize the distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='primary_focus_subject', data=new_df, order=new_df['primary_focus_subject'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Primary Focus Subject')
plt.xlabel('Primary Focus Subject')
plt.ylabel('Count')
plt.show()
```

```python
# Check for missing values in the 'resource_type' column
print(new_df['resource_type'].isnull().sum())

# Fill missing values with a placeholder (or remove rows with missing values)
new_df['resource_type'].fillna('Missing', inplace=True)

# Display the distribution
print(new_df['resource_type'].value_counts())

# Visualize the distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='resource_type', data=new_df, order=new_df['resource_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Distribution of Resource Type')
plt.xlabel('Resource Type')
plt.ylabel('Count')
plt.show()
```
